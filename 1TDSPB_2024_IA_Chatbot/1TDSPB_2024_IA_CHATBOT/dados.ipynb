# Aqui estamos analisando como as entidades do mundo afetaram com a Produção média anual de plástico

```python
# Carregar os dados
file_path = '/mnt/data/1- producao-de-plastico-global.csv'
data = pd.read_csv("1- producao-de-plastico-global.csv")

# Agrupar por 'Entidade' e calcular a produção média anual de plástico
mean_production = data.groupby('Entidade')['Produção Anual de Plástico'].mean().reset_index()

# Ordenar os dados por produção média para melhor visualização
mean_production = mean_production.sort_values(by='Produção Anual de Plástico', ascending=False)

# Configurar o estilo do gráfico
sns.set(style="whitegrid")

# Criar o gráfico de barras
plt.figure(figsize=(12, 8))
sns.barplot(x='Produção Anual de Plástico', y='Entidade', data=mean_production, palette='viridis')

# Adicionar título e rótulos aos eixos
plt.title('Produção Média Anual de Plástico por Entidade')
plt.xlabel('Produção Anual de Plástico')
plt.ylabel('Entidade')

# Exibir o gráfico
plt.show()
```

# Seguindo a mesma linha de raciocínio aqui estamos analisando como as entidades do mundo afetaram com a Participação na emissão global de plásticos para o oceano

# E por fim conseguimos ver a relação Lixo plástico mal gerenciado por pessoa (kg por ano) com as Entidades(Países)

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Carregar os dados
file_path4 = '4- desperdicio-plastico-per-capita.csv'
data = pd.read_csv(file_path4)

# Remover espaços extras nos nomes das colunas
data.columns = data.columns.str.strip()

# Configurar o estilo do gráfico
sns.set(style="whitegrid")

# Criar o gráfico de linha
plt.figure(figsize=(12, 60))
sns.lineplot(x='Lixo plástico mal gerenciado por pessoa (kg por ano)', y='Ano', hue='Entidade', data=data, marker='o')

# Adicionar título e rótulos aos eixos
plt.title('Lixo plástico mal gerenciado por pessoa ao longo do tempo')
plt.xlabel('Lixo plástico mal gerenciado por pessoa (kg por ano)')
plt.ylabel('Lixo plástico mal gerenciado por pessoa (kg por ano)')



# Exibir o gráfico
plt.legend(title='Entidade', loc='upper left', bbox_to_anchor=(1, 1))
plt.tight_layout()
plt.show()
```

## transformando strings em int

```python
# Criar um mapeamento das strings para inteiros
unique_entities = df_arquivos_final[' Entidade'].unique()
entity_to_int = {entity: i for i, entity in enumerate(sorted(unique_entities))}

# Aplicar o mapeamento à coluna "Entidade"
df_arquivos_final[' Entidade'] = df_arquivos_final[' Entidade'].map(entity_to_int)

df_treino1 = df_arquivos_final


```

```python
df_treino1.sample(40)
```

### Aqui fizemos uma rápida análise da Poluição da Água com as outras colunas

## Enfim o treinamento da máquina

```python
X = df_treino1.drop(" Entidade", axis=1)
y = df_treino1[" Entidade"]
```

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
```

```python
dc_tree = DecisionTreeClassifier()
dc_tree.fit(X, y)
```

```python
plot_tree(dc_tree);
```

```python
from sklearn.metrics import accuracy_score
```

```python
from sklearn.metrics import accuracy_score

y_pred = dc_tree.predict(X)
accuracy_score(y, y_pred)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.30, random_state=20)
```

```python
tree_randomv2 = DecisionTreeClassifier()
tree_randomv2.fit(X_train, y_train)
tree_randomv2
```

```python
print(accuracy_score(y_train, tree_randomv2.predict(X_train)))
print(accuracy_score(y_test, tree_randomv2.predict(X_test)))
```

```python
from sklearn.ensemble import RandomForestClassifier

floresta = RandomForestClassifier().fit(X_train, y_train)
```

```python
y_pred = floresta.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test, y_pred)*100
```
